```python
import sympy as sy
from sympy import exp, log, sqrt

def norm(x):
    d = x.shape[0]
    n = 0
    for i in range(d):
        n += x[i,0]**2
    return sy.sqrt(n)
```

```python
# アトラクター
x_1, x_2, x_3 = sy.symbols('x_1, x_2, x_3')
x = sy.Matrix([[x_1, x_2, x_3]]).T
v_1, v_2, v_3 = sy.symbols('v_1, v_2, v_3')
v = sy.Matrix([[v_1, v_2, v_3]]).T


# x = sy.MatrixSymbol('x',  3, 1)
# v = sy.MatrixSymbol('v',  3, 1)

sigma_alpha, sigma_gamma, epsilon, w_u, w_l, k, eta = \
    sy.symbols('sigma_alpha, sigma_gamma, epsilon, w_u, w_l, k, eta')


alpha = exp(-norm(x)**2 / (2*sigma_alpha**2))
gamma = exp(-norm(x)**2 / (2*sigma_gamma**2))
w = gamma*w_u + (1-gamma)*w_l
grad_psi_1 = k * (1-exp(-2*eta*norm(x)))/(1+exp(-2*eta*norm(x)))*x/norm(x)

Le = w * v.T * ((1-alpha)*grad_psi_1*grad_psi_1.T + (alpha+epsilon)*sy.eye(3)) * v
Le
```

```python
print(sy.diff(sy.expand(Le), x[0,0]))
```

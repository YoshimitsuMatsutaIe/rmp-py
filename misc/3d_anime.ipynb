```python
import matplotlib.pyplot as plt
import matplotlib.animation as anm
from IPython.display import HTML
import matplotlib
matplotlib.rcParams['savefig.facecolor'] = 'white'
import numpy as np
```

```python
x = []
y = []
for i in range(10):
    x.append(np.arange(0, np.pi, 0.1))
    y.append(np.cos(x[i]+np.pi/6*i))
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
p = ax.scatter(x[0], y[0])
time_template = 'time = %d [i]'
tx = ax.set_title(time_template % 0)

def update(i):
    temp = np.stack([x[i], y[i]])
    p.set_offsets(temp.T)
    tx.set_text(i)
    return p, tx

ani = anm.FuncAnimation(
    fig = fig,
    func = update,
    frames=range(len(x)),
    blit=True
)
HTML(ani.to_jshtml())
```

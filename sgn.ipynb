```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1, 1, 0.01)
```

```python
def sgn(x_dot):
    if x_dot < 0:
        return 1
    else:
        return 0

y_sgn = [sgn(i) for i in x]
plt.plot(x, y_sgn)
plt.grid()
plt.axis('equal')
```

```python
def sigmoid(x, alpha):
    return 1 / (1 + sy.exp(-alpha * x))

y_sig = [-sigmoid(i, 10)+1 for i in x]
plt.plot(x, y_sig)
plt.grid()
plt.axis('equal')
```

```python
def s2(x, r, alpha,):
    return 1/2 * (np.tanh(-alpha*(abs(x)-r)) + 1)

y_s2 = [-s2(i, 0.5, 10)+1 for i in x]
plt.plot(x, y_s2)
plt.grid()
plt.axis('equal')
```
